```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
low_memory = False
```

```python
total = []
nainfo = []
naraca = []
branco = []
negro = []
pardo = []
amarelo = []
indigena = []
arquivo = ['enem2014.csv','enem2015.csv']
arquivo2 = ['enem2016.csv','enem2017.csv','enem2018.csv']
anos = ['2014','2015','2016','2017','2018']
```

```python
for i in arquivo:
    df = pd.read_csv(i, usecols=['TP_COR_RACA'], delimiter = ',',encoding = 'iso-8859-1')
    naraca.append(df.query('(TP_COR_RACA == 0)')['TP_COR_RACA'].count())
    branco.append(df.query('(TP_COR_RACA == 1)')['TP_COR_RACA'].count())
    negro.append(df.query('(TP_COR_RACA == 2)')['TP_COR_RACA'].count())
    pardo.append(df.query('(TP_COR_RACA == 3)')['TP_COR_RACA'].count())
    amarelo.append(df.query('(TP_COR_RACA == 4)')['TP_COR_RACA'].count())
    indigena.append(df.query('(TP_COR_RACA == 5)')['TP_COR_RACA'].count())
    nainfo.append(df.query('(TP_COR_RACA == 6)')['TP_COR_RACA'].count())
    total.append(df.query('(TP_COR_RACA)')['TP_COR_RACA'].count())
```

```python
for i in arquivo2:
    df = pd.read_csv(i, usecols=['TP_COR_RACA'], delimiter = ';',encoding = 'iso-8859-1')
    naraca.append(df.query('(TP_COR_RACA == 0)')['TP_COR_RACA'].count())
    branco.append(df.query('(TP_COR_RACA == 1)')['TP_COR_RACA'].count())
    negro.append(df.query('(TP_COR_RACA == 2)')['TP_COR_RACA'].count())
    pardo.append(df.query('(TP_COR_RACA == 3)')['TP_COR_RACA'].count())
    amarelo.append(df.query('(TP_COR_RACA == 4)')['TP_COR_RACA'].count())
    indigena.append(df.query('(TP_COR_RACA == 5)')['TP_COR_RACA'].count())
    nainfo.append(df.query('(TP_COR_RACA == 6)')['TP_COR_RACA'].count())
    total.append(df.query('(TP_COR_RACA)')['TP_COR_RACA'].count())
```

```python
fig, ax = plt.subplots()
plt.rcParams['figure.figsize'] = (20,10)
ax.plot(anos, total, label='Total')
ax.plot(anos, naraca, label='No declarado')
ax.plot(anos, branco, label='Branco')
ax.plot(anos, negro, label='Negro')
ax.plot(anos, pardo, label='Pardo')
ax.plot(anos, amarelo, label='Amarelo')
ax.plot(anos, indigena, label='Indgena')
ax.plot(anos, nainfo, label='No dispe da informao')
ax.grid()
plt.legend()
plt.title('Inscritos no Enem perodo de 2014 a 2018 com base em suas Etnias ')
plt.xlabel('Anos')
plt.ylabel('Quantidade')
plt.show()
```
